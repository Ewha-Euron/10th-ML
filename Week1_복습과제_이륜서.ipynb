### 15. What is the result of the following expression?

(실행 후, 결과에 대한 주석 작성해주세요. ex. 출력 결과에 대한 이유)

```python
0 * np.nan
np.nan == np.nan
np.inf > np.nan
np.nan - np.nan
np.nan in set([np.nan])
0.3 == 3 * 0.1
```

````python
print("0 * np.nan =",0 * np.nan)
#nan은 알 수 없는 값이기 때문에 0을 곱해도 여전히 알 수 없는 값이다.

print("np.nan == np.nan =",np.nan == np.nan)
#첫 번째와 마찬가지로 nan은 '알 수 없는 값'이기 때문에 자기 자신과도 비교할 수 없다.

print("np.inf > np.nan =",np.inf > np.nan)
#inf는 무한을 나타냄. nan은 정의할 수 없으므로 어떤 값과도 비교할 수 없다.

print("np.nan - np.nan =",np.nan - np.nan)
#알 수 없는 값에서 알 수 없는 값을 빼기 때문에 결과도 알 수 없다

print("np.nan in set([np.nan]) =",np.nan in set([np.nan]))
#set : 집합을 의미. { }를 사용, 중복을 허용X, 순서 X  /in : 어떤 값이 안에 있는지 확인
#set 안에 같은 객체인 nan이 있으므로 True이다. 데이터의 실제성을 물어봄

print("0.3 == 3 * 0.1 =",0.3 == 3 * 0.1)
#부동 소수점의 오차 문제로 0.1을 정확하게 표현할 수 없어서 두 값은 같지 않다.
````

### 21. Given a DataFrame of random numeric values:
```python
df = pd.DataFrame(np.random.random(size=(5, 3))) # this is a 5x3 DataFrame of float values
```

how do you subtract the row mean from each element in the row?

````python
# 랜덤 시드 고정
np.random.seed(2025)

df = pd.DataFrame(np.random.random(size=(5, 3)))

df.apply(lambda x : x-x.mean(),axis=1)
````
