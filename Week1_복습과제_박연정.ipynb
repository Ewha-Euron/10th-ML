### 15. What is the result of the following expression?

(실행 후, 결과에 대한 주석 작성해주세요. ex. 출력 결과에 대한 이유)

```python
0 * np.nan
np.nan == np.nan
np.inf > np.nan
np.nan - np.nan
np.nan in set([np.nan])
0.3 == 3 * 0.1
```

````python
# NaN은 정의되지 않은 값이라 어떤 수와 연산해도 결과가 NaN
0 * np.nan
````

````python
# NaN은 어떤 값과도 같다고 판단하지 않으므로 False
# 자기 자신과 비교해도 False
np.nan == np.nan
````

````python
# NaN은 비교 자체가 불가능하므로 False
np.inf > np.nan
````

````python
# 제일 처음과 마찬가지로 정의되지 않은 값이라 어떤 수와 연산해도 NaN
np.nan - np.nan
````

### 18. What are the result of the following expressions?

(실행 후, 결과에 대한 주석 작성해주세요. ex. 출력 결과에 대한 이유)

```python
np.array(0) / np.array(0)
np.array(0) // np.array(0)
np.array([np.nan]).astype(int).astype(float)
```

````python
# 0/0은 정의되지 않음
# 결과 값은 NaN
np.array(0) / np.array(0)
````

````python
# //은 정수 나눗셈
# 0//0은 수학적으로 정의되지 않는 연산이라 경고가 뜸
# 결과 값은 0
np.array(0) // np.array(0)
````

````python
# NaN은 정수로 변활될 수 없으므로
# int64의 가장 작은 값으로 대체해서 처리
# 해당 숫자를 float으로 바꾸면 결과 값이 나옴
np.array([np.nan]).astype(int).astype(float)
````
