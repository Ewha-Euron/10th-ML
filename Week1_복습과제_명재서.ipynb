### 15. What is the result of the following expression?

(실행 후, 결과에 대한 주석 작성해주세요. ex. 출력 결과에 대한 이유)

```python
0 * np.nan
np.nan == np.nan
np.inf > np.nan
np.nan - np.nan
np.nan in set([np.nan])
0.3 == 3 * 0.1
```

````python
print(0 * np.nan) # nan은 정의되지 않은 값으로, 어떤 수를 곱해도 결과는 nan임,
print(np.nan == np.nan) # nan은 자기 자신과도 '같지 않다'고 정의되어 있습니다.
print(np.inf > np.nan) # nan은 숫자가 아니므로 무한대(inf)와 크기를 비교할 수 없으며, 결과는 항상 False입니다.
print(np.nan - np.nan) #정의되지 않은 값끼리의 연산은 결과 또한 정의할 수 없으므로 nan이 나옵니다.
print(np.nan in set([np.nan])) # set(집합)이나 list 내부에 동일한 객체(메모리 주소)로 존재하는지 확인하는 것이므로 True가 나옵니다.
print(0.3 == 3 * 0.1) #set(집합)이나 list 내부에 동일한 객체(메모리 주소)로 존재하는지 확인하는 것이므로 True가 나옵니다.
````

### 21. Given a DataFrame of random numeric values:
```python
df = pd.DataFrame(np.random.random(size=(5, 3))) # this is a 5x3 DataFrame of float values
```

how do you subtract the row mean from each element in the row?

````python
# 랜덤 시드 고정
np.random.seed(2025)

df = pd.DataFrame(np.random.random(size=(5, 3)))


````

````python
df = df.sub(df.mean(axis=1), axis=0)

print(df)
````
