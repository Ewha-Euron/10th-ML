```python
# 문제 풀이 전 numpy 임포트
import numpy as np

# 3. Create a DataFrame df from this dictionary data which has the index labels

data = {
    'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
    'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
    'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']
}

labels = ['a','b','c','d','e','f','g','h','i','j']

df = pd.DataFrame(data, index=labels)

df
```

```python
# 4. Display a summary of the basic information about this DataFrame
df.info()
```

```python
# 5. Display a summary of the basic statistics about data of this DataFrame
df.describe()
```

```python
# 6. Return the first 3 rows
df.head(3)
```

```python
# 7. Select just the 'animal' and 'age' columns
df[['animal','age']]
```

```python
# 8. Select rows [3,4,8] and columns ['animal','age']
df.loc[df.index[[3,4,8]], ['animal','age']]
```

```python
# 9. Select rows where visits > 3
df[df['visits'] > 3]
```

```python
# 10. Select rows where age is NaN
df[df['age'].isna()]
```

```python
# 11. animal == cat AND age < 3
df[(df['animal']=='cat') & (df['age'] < 3)]
```

```python
# 12. age between 2 and 4
df[df['age'].between(2,4)]
```

```python
# 13. Change age in row 'f' to 1.5
df.loc['f','age'] = 1.5
df
```

```python
# 14. sum of visits
df['visits'].sum()
```

```python
# 15. mean age for each animal
df.groupby('animal')['age'].mean()
```

```python
# 16. append row 'k' then delete it

df.loc['k'] = ['dog', 4, 1, 'yes']
print("After adding row k")
print(df)

df = df.drop('k')
print("\nAfter deleting row k")
print(df)
```

```python
# 17. count each animal
df['animal'].value_counts()
```

```python
# 18. Sort by age(desc), visits(asc)

df_sorted = df.sort_values(by=['age','visits'], ascending=[False, True])
df_sorted
```

```python
# 19. replace yes/no with boolean

df['priority'] = df['priority'].map({'yes': True, 'no': False})
df
```

```python
# 20. change snake to python

df['animal'] = df['animal'].replace('snake','python')
df
```

```python
# 랜덤 시드 고정
np.random.seed(2025)

df = pd.DataFrame(np.random.random(size=(5, 3)))
print("original df")
print(df)

# subtract row mean
df_sub = df.sub(df.mean(axis=1), axis=0)

print("\nrow mean subtracted")
print(df_sub)
```
