### 15. What is the result of the following expression?

(실행 후, 결과에 대한 주석 작성해주세요. ex. 출력 결과에 대한 이유)

```python
0 * np.nan
np.nan == np.nan
np.inf > np.nan
np.nan - np.nan
np.nan in set([np.nan])
0.3 == 3 * 0.1
```

````python
print(0 * np.nan)
# 결측값이 계산 과정에서 사라지는 것을 방지하기 위해, 연산에 NaN이 포함되면 결과도 NaN으로 유지된다.

print(np.nan == np.nan)
# NaN은 '알 수 없는 값'을 의미하기 때문에 자기 자신과도 같다고 판단하지 않는다. 

print(np.inf > np.nan)
# NaN은 정의되지 않은 값이므로 어떤 값과도 크기 비교가 성립하지 않는다.

print(np.nan - np.nan)
# 연산에 NaN이 포함되면 결과도 NaN이다. 

print(np.nan in set([np.nan]))
# set 안에 같은 객체의 NaN이 존재하기 때문에 True가 반환된다.

print(0.3 == 3 * 0.1)
# 부동소수점 표현의 정밀도 문제로 0.1을 정확하게 표현할 수 없어 3*0.1의 계산 결과가 0.3과 정확히 같지 않기 때문이다. 
````

### 21. Given a DataFrame of random numeric values:
```python
df = pd.DataFrame(np.random.random(size=(5, 3))) # this is a 5x3 DataFrame of float values
```

how do you subtract the row mean from each element in the row?

````python
# 랜덤 시드 고정
np.random.seed(2025)

df = pd.DataFrame(np.random.random(size=(5, 3)))

df.apply(lambda x : x-x.mean(),axis=1)
````
