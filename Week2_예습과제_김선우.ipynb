```python
from sklearn.base import BaseEstimator

class MyDummyClassifier (BaseEstimator):
    def fit(self, X, y=None):
        pass
    def predict(self, X):
        pred = np.zeros( (X.shape[0],1))
        for i in range (X.shape[0]):
            if X['sex'].iloc[i] == 1: # Changed 'Sex' to 'sex'
                pred[i] = 0
            else:
                pred[i] = 1
        return pred
```

```python
# 2.6

def fillna(df):
    df['age'].fillna(df['age'].mean(), inplace=True)
    # Explicitly convert 'deck' to object type before filling NaNs to avoid TypeError with Categorical
    df['deck'] = df['deck'].astype(object)
    df['deck'].fillna('N', inplace=True)
    df['embarked'].fillna('N', inplace=True)
    df['fare'].fillna(0, inplace=True)
    return df

def drop_features(df):
    # The seaborn titanic dataset does not have 'PassengerId', 'Name', 'Ticket' in X_titanic_df directly
    # We drop similar irrelevant columns from the seaborn dataset if they exist
    columns_to_drop = ['class', 'who', 'adult_male', 'embark_town', 'alive', 'alone', 'pclass', 'sibsp', 'parch']
    for col in columns_to_drop:
        if col in df.columns:
            df.drop([col], axis=1, inplace=True)
    return df

def format_features(df):
    # Convert 'deck' to string type before taking the first character
    if 'deck' in df.columns:
        df['deck'] = df['deck'].astype(str).str[:1]

    # Features to Label Encode
    features = ['deck', 'sex', 'embarked']
    for feature in features:
        if feature in df.columns:
            le = LabelEncoder()
            le = le.fit(df[feature])
            df[feature] = le.transform(df[feature])
    return df

def transform_features(df):
    df = fillna(df)
    df = drop_features(df)
    df = format_features(df)
    return df
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import seaborn as sns
from sklearn.preprocessing import LabelEncoder # LabelEncoder가 필요합니다.

# Load the titanic dataset from seaborn
df = sns.load_dataset('titanic')
print(df.head())

# Separate target variable (survived) and features
y_titanic_df = df['survived']
X_titanic_df = df.drop('survived', axis=1)

# --- Data Preprocessing for the seaborn titanic dataset ---
# Missing value handling (corresponding to original fillna logic)
X_titanic_df['age'].fillna(X_titanic_df['age'].mean(), inplace=True)
# Convert 'deck' to object type before filling NaNs to avoid TypeError with Categorical
X_titanic_df['deck'] = X_titanic_df['deck'].astype(object)
X_titanic_df['deck'].fillna('N', inplace=True) # 'deck' column corresponds to 'Cabin'
X_titanic_df['embarked'].fillna('N', inplace=True)
X_titanic_df['fare'].fillna(0, inplace=True) # 'fare' column is already numeric, just handle NaNs

# Format 'deck' and Label Encode categorical features
X_titanic_df['deck'] = X_titanic_df['deck'].astype(str).str[:1] # Take first letter for 'deck'

# Categorical features for Label Encoding (corresponding to original format_features logic)
categorical_features = ['sex', 'embarked', 'deck'] # 'sex' corresponds to 'Sex', 'embarked' to 'Embarked'

for feature in categorical_features:
    le = LabelEncoder()
    # Check if the feature exists before encoding
    if feature in X_titanic_df.columns:
        X_titanic_df[feature] = le.fit_transform(X_titanic_df[feature])

# Drop irrelevant columns from the seaborn dataset if they are not useful for the model
# The original transform_features dropped 'PassengerId', 'Name', 'Ticket', which are not in seaborn df.
# We will drop similar redundant or irrelevant columns from the seaborn df.
columns_to_drop = ['class', 'who', 'adult_male', 'embark_town', 'alive', 'alone']
X_titanic_df.drop(columns_to_drop, axis=1, inplace=True, errors='ignore') # 'errors='ignore' prevents errors if a column is already gone

# Now, X_titanic_df is processed.
X_train, X_test, y_train, y_test = train_test_split(X_titanic_df, y_titanic_df,
                                                    test_size=0.2, random_state=0)

myclf = MyDummyClassifier()
myclf.fit(X_train, y_train)
mypredictions = myclf.predict(X_test)
print('Dummy Classifier의 정확도는: {0:4f}'.format(accuracy_score(y_test, mypredictions)))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix

def get_clf_eval(y_test, pred):
    confusion = confusion_matrix(y_test, pred)
    accuracy = accuracy_score(y_test, pred)
    precision = precision_score(y_test, pred)
    recall = recall_score(y_test, pred)
    print('오차 행렬')
    print(confusion)
    print('정확도: {0:.4f}, 정밀도: {1:.4f}, 재현율: {2:.4f}'.format(accuracy, precision, recall))
```
