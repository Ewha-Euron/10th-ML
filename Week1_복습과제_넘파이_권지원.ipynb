```python
# =========================================
# 1. Import the numpy package under the name np.
# =========================================
import numpy as np
```

```python
# =========================================
# 2. Print the numpy version and the configuration.
# =========================================
print("NumPy version:", np.__version__)
np.show_config()
```

```python
# =========================================
# 3. Create a null vector of size 10.
# =========================================
print("\n3번")
v = np.zeros(10)
print(v)
```

```python
# =========================================
# 4. Create a null vector of size 10 but the fifth value which is 1.
# =========================================
print("\n4번")
v = np.zeros(10)
v[4] = 1
print(v)
```

```python
# =========================================
# 5. Create a vector with values ranging from 10 to 49.
# =========================================
print("\n5번")
v = np.arange(10, 50)
print(v)
```

```python
# =========================================
# 6. Reverse a vector (first element becomes last).
# =========================================
print("\n6번")
v = np.arange(10)
print(v[::-1])
```

```python
# =========================================
# 7. Create a 3x3 matrix with values ranging from 0 to 8.
# =========================================
print("\n7번")
m = np.arange(9).reshape(3,3)
print(m)
```

```python
# =========================================
# 8. Find indices of non-zero elements from [1,2,0,0,4,0].
# =========================================
print("\n8번")
arr = np.array([1,2,0,0,4,0])
print(np.nonzero(arr))
```

```python
# =========================================
# 9. Create a 3x3 identity matrix.
# =========================================
print("\n9번")
print(np.eye(3))
```

```python
# =========================================
# 10. Create a 3x3x3 array with random values.
# =========================================
print("\n10번")
print(np.random.random((3,3,3)))
```

```python
# =========================================
# 11. Create a 10x10 array with random values and find the minimum and maximum values.
# =========================================
print("\n11번")
arr = np.random.random((10,10))
print(arr.min(), arr.max())
```

```python
# =========================================
# 12. Create a random vector of size 30 and find the mean value.
# =========================================
print("\n12번")
arr = np.random.random(30)
print(arr.mean())
```

```python
# =========================================
# 13. Create a 2d array with 1 on the border and 0 inside. (size: 10x10)
# =========================================
print("\n13번")
arr = np.ones((10,10))
arr[1:-1,1:-1] = 0
print(arr)
```

```python
# =========================================
# 14. Add a border filled with 0's around an existing array
# =========================================
print("\n14번")

# before
arr = np.ones((5,5))
print("before")
print(arr)

# after
arr[1:-1,1:-1] = 1
arr[0,:] = 0
arr[-1,:] = 0
arr[:,0] = 0
arr[:,-1] = 0

print("\nafter")
print(arr)
```

```python
# =========================================
# 15. Expression results with explanation
# =========================================
print("\n15번")

print(0 * np.nan)
# nan : NaN이 포함된 연산 결과는 대부분 NaN이 됨

print(np.nan == np.nan)
# False : NaN은 자기 자신과도 같지 않다고 정의됨

print(np.inf > np.nan)
# False : NaN이 포함된 비교 연산은 항상 False

print(np.nan - np.nan)
# nan : NaN 연산 결과는 NaN

print(np.nan in set([np.nan]))
# True : set 안에 동일한 객체 NaN이 존재하므로 True

print(0.3 == 3 * 0.1)
# False : 부동소수점 정밀도 문제
```

```python
# =========================================
# 16. Normalize a 5x5 random matrix
# =========================================
print("\n16번")

arr = np.random.random((5,5))
normalized = (arr - arr.mean()) / arr.std()
print(normalized)
```

```python
# =========================================
# 17. Multiply a 5x3 matrix by a 3x2 matrix
# =========================================
print("\n17번")

A = np.ones((5,3))
B = np.ones((3,2))

print(np.dot(A,B))
```

```python
# =========================================
# 18. Expression results with explanation
# =========================================
print("\n18번")

print(np.array(0) / np.array(0))
# nan : 0/0 은 정의되지 않아 NaN

print(np.array(0) // np.array(0))
# 0 : 정수 floor division에서는 0으로 처리됨 (warning 발생 가능)

print(np.array([np.nan]).astype(int).astype(float))
# 매우 큰 음수값으로 변환됨 : NaN을 int로 캐스팅할 때 표현 가능한 최소 정수로 변환됨
```

```python
# =========================================
# 19. Create a 5x5 matrix with row values ranging from 0 to 4
# =========================================
print("\n19번")

arr = np.tile(np.arange(5), (5,1))
print(arr.astype(float))
```

```python
# =========================================
# 20. Create a random vector of size 10 and sort it
# =========================================
print("\n20번")

arr = np.random.random(10)
arr.sort()
print(arr)
```
